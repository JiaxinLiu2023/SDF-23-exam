```python
for i in range(1, 10):
    if i % 2 == 0:#余数公式
        print("%d is even" % i)#它告诉Python将后面的整数 i 的值插入到这个位置
    elif i % 3 == 0:#否则
        print("%d is multiple of 3" % i)
    else:#否则
        print("%d is odd" % i)
```

```python
import matplotlib.pyplot as plt
np.random.seed(1000)
y = np.random.standard_normal(20)#20个随机数
x = np.arange(len(y))#len(y) 是一个Python内置函数，用于返回一个数据结构（比如列表、元组、字符串等）中元素的数量，即数据结构的长度。
plt.figure(figsize=(10, 6))#尺寸
plt.plot(x, y)#plt.plot() 函数用于绘制折线图，其中 x 是横坐标，y 是纵坐标
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```
